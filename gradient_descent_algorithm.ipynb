# Algorithm of gradient descent using Python

a = 0.2<br>
b = 0.4<br>
c = 0.9<br>
η = 0.4

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Parameters:

```python
a = 0.2
b = 0.4
c = 0.9

r = np.array([a, b])
R = np.array([[1, c], [c, 1]])

η = 0.4
```

Cost function:

```python
def E(w, r, R, σ):
    return 0.5 * σ ** 2 - np.dot(r, w) + 0.5 * np.dot(np.dot(w, R), w)
```

A function that calculates the value of the vector of weights using the gradient descent method:

```python
def gradient_descent(n_iter, r, R, η, σ):
    weights = []
    costs = []
    
    w_0 = np.array([1, 1])
    weights.append(w_0)
    c_0 = E(w_0, r, R, σ)
    costs.append(c_0)
    for i in range(n_iter):
        w = weights[i] - η * (np.dot(R, weights[i]) - r)
        weights.append(w)
        c = E(w, r, R, σ)
        costs.append(c)

    return weights, costs
```

Calculation of the value of w(t), t = 1, 2, ..., T at a given value of the parameter η using the gradient descent method

```python
weights, costs = gradient_descent(100, r, R, η, 0)
df = pd.DataFrame(weights, columns=["x", "y"])
df["E(w)"] = costs
df
```

Plotting the trajectory of the weight vector w(t) in the W plane

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.scatter(df["x"], df["y"])
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection="3d")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("E(w)")
ax.plot(df["x"], df["y"], df["E(w)"])
plt.show()
```

Selection of the learning rate η1, at which the gradient descent algorithm coincides and plotting the weight trajectory

```python
η1 = 0.008
weights_conv, costs_conv = gradient_descent(100, r, R, η1, 0)
df_conv = pd.DataFrame(weights_conv, columns=["x", "y"])
df_conv["E(w)"] = costs_conv
df_conv
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.scatter(df_conv["x"], df_conv["y"])
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection="3d")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("E(w)")
ax.plot(df_conv["x"], df_conv["y"], df_conv["E(w)"])
plt.show()
```

Selection of the learning rate η2, at which the gradient descent algorithm diverges and plotting the weight trajectory

```python
η2 = 1.1
weights_div, costs_div = gradient_descent(100, r, R, η2, 0)
df_div = pd.DataFrame(weights_div, columns=["x", "y"])
df_div["E(w)"] = costs_div
df_div
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.scatter(df_div["x"], df_div["y"])
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection="3d")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("E(w)")
ax.plot(df_div["x"], df_div["y"], df_div["E(w)"])
plt.show()
```
